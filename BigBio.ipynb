```python
from opentree import OT
```

Get the list of taxa from a textfile.  

```python
fi = open("bigbio_names.txt").readlines()
```

```python
fi[0].split(';')
```

```python
fi[0].split(';')[5]
```

```python
all_genera = set()
for taxon in fi:
    try:
        genus = taxon.split(';')[5]
        all_genera.add(genus)
    except:
        pass
        
```

```python
len(all_genera)
```

```python
all_genera
```

```python
OT.get_ottid_from_name('Stappia')
OT.get_ottid_from_name('Fuscheria')
```

```python
matches = dict()
all_ottids = set()
for genus in all_genera:
    if len(genus) > 2:
        print(genus)
        ott_id = OT.get_ottid_from_name(genus)
        print(ott_id)
        all_ottids.add(ott_id)
        matches[genus] = ott_id
```

```python
matches.values()
```

```python
output = OT.synth_induced_tree(ott_ids=list(matches.values()),  label_format='name_and_id')
```

```python
print(output)
```

```python
output.response_dict['broken']
```

```python
treefile = "BigBio.tre"
output.tree.write(path = treefile, schema = "newick")
output.tree.print_plot(width=100)
```

```python
studies = output.response_dict['supporting_studies']
cites = OT.get_citations(studies) #this can be a bit slow
print(cites)
```
